```python
import pandas as pd 
import seaborn as sns
data = pd.read_csv("energy.csv")
data.head(10)
```

```python
sample_data = data.sample(10)
sample_data
```

```python
sns.catplot(x='Year', y='Megawatthours', data = sample_data)
```

```python
sns.lmplot(x="Year", y="Megawatthours",data=sample_data)
```

```python
sns.barplot(x = 'Year', y = "Megawatthours", hue = 'State', data = sample_data)
```

```python
g = data[(data["Energy Source"]=="Coal")& (data["Year"] <1995)]
g
```

```python
sns.catplot(x='Year', y='Megawatthours', hue = 'Energy Source', kind = 'swarm', data = g)
```

```python
f = data[(data["State"]=="NY") &(data["Year"] <2000)]
f
```

```python
sns.catplot(x='Year', y='Megawatthours', kind = 'bar', data = f)
```

```python
sns.scatterplot(x='Year', y='Megawatthours', data=f)
```

```python
h = data[data['State'].isin(['AL', 'NY', 'IL', 'SD', 'LA']) & (data['Year'] < 2000)]
h
```

```python
sns.catplot(x='Year', y='Megawatthours', hue='State', data=h)
```
